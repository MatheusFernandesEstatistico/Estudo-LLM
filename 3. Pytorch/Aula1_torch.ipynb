## 1. Introdução aos *Tensores*

Tensores são o bloco de construção fundamental do aprendizado de máquina. Sua função é representar dados de forma numérica. Por exemplo, você pode representar uma imagem como um tensor com formato [3, 224, 224], que significa [canais_de_cor, altura, largura], ou seja, a imagem tem 3 canais de cor (vermelho, verde, azul), uma altura de 224 pixels e uma largura de 224 pixels.
